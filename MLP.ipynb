```python
from keras.datasets.mnist import load_data
#((訓練圖片, 訓練答案), (測試圖片, 測試答案))
(x_train, y_train), (x_test, y_test) = load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
c = random.randint(0, 5999)
plt.title("This is "+str(y_train[c]))
plt.axis("off") #關閉軸座標
plt.imshow(x_train[c], cmap="gray")
```

```python
import pandas as pd
pd.DataFrame(x_train[c]) #以數字表示圖形
```

```python
from keras.models import Sequential
from keras.layers import Dense
mlp = Sequential()
#784 * 256 + 256(bias)
mlp.add(Dense(256,activation="relu",input_dim=784)) #中間層relu
#256 * 10 + 10(bias)
mlp.add(Dense(10,activation="softmax")) #多個softmax
mlp.summary()
```

```python
mlp.compile(loss = "categorical_crossentropy", #參數單個（距離）
            optimizer = "adam", #adam優化動量、步長
           metrics = ["accuracy"])
```

```python
#對圖片處理：降維（1D） -> 0~1（開始的隨機weight才能配合）
x_train_norm = x_train.reshape(60000,784) / 255 #全部都會/255
x_test_norm = x_test.reshape(10000,784) / 255
```

```python
#對答案處理：One-Hot Encoding -> 10個機率
from keras.utils.np_utils import to_categorical
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
print("Before : ", y_train[0])
print("After : ", y_train_cat[0])
```

```python
#batch_size：依資料筆數調整（梯度下降）一次
#資料小（100-200），資料大（10-20）
#調整次數：60000 * 0.9 * 10(epochs) / 200(batch) 
mlp.fit(x=x_train_norm,y=y_train_cat,
        batch_size = 200, #資料小
        epochs=10, #整份資料(60000)幾次[模型停止訓練時機]
        validation_split=0.1, #用新資料驗證模型好壞
        verbose=2) #log多寡（1詳盡；2必要；0不顯示）
#val_loss訓練到平滑時為最好階段（停止） -> 過度訓練
```

```python
mlp.evaluate(x_test_norm, y_test_cat)
```

```python
from sklearn.metrics import confusion_matrix
base = list(range(10))
i = [str(b) + "(原本)" for b in base]
c = [str(b) + "(預測)" for b in base]

pre = mlp.predict_classes(x_test_norm)
pd.DataFrame(confusion_matrix(y_test, pre),
             columns = c,
             index = i)
```

```python
import numpy as np
idx = np.nonzero(pre != y_test)[0]
false_img = x_test[idx] #預測錯誤的圖片
false_answer = y_test[idx]#預測錯誤圖片的正確答案
false_pre = pre[idx] #預測錯誤圖片的預測標籤

width = 10
height = len(false_img) // width + 1
plt.figure(figsize=(14,40))
for i,img in enumerate(false_img):
    plt.subplot(height, width, i+1) #subplot(高, 寬, 張)
    t = "[O]" + str(false_answer[i]) + "[P]" + str(false_pre[i]) #O->original
    plt.title(t)
    plt.axis("off")
    plt.imshow(img, cmap="gray")
```
